```python
# Publish Table to Jupyter Notebook
# 
# This snippet is configured to load <200> rows of Xcalar table <yelp_academic_dataset_user_fixed#Zz1> into a pandas dataframe named
# <yelp_academic_dataset_user_fixed#Zz1_pd>.
#
# To instantiate or refresh your pandas dataframe, run the Connect snippet, 
# and then run this snippet. 
#
# Best Practice is not to edit this code. 
#
# To use different data with this Jupyter Notebook:
# 1) Go to the table in your Xcalar Workbook.
# 2) From the table menu, click Publish to Jupyter.
# 3) Click full table or enter a number of rows and click submit.

# Imports data into a pandas dataframe.
def getDataFrameFromDict():
    from collections import OrderedDict
    resultSetPtr_Zz1 = ResultSet(xcalarApi, tableName="yelp_academic_dataset_user_fixed#Zz1", maxRecords=200)
    yelp_academic_dataset_user_fixed_Zz1 = []
    for row in resultSetPtr_Zz1:
        col_list = ["user_id",]
        kv_list = []
        for k in col_list:
            if k not in row:
                kv_list.append((k, None))
            else:
                kv_list.append((k, row[k]))
                if type(row[k]) is list:
                    for i in range(len(row[k])):
                        subKey = k + "[" + str(i) + "]"
                        if subKey in col_list:
                            row[subKey] = row[k][i]
        filtered_row = OrderedDict(kv_list)

        yelp_academic_dataset_user_fixed_Zz1.append(filtered_row)
    return pd.DataFrame.from_dict(yelp_academic_dataset_user_fixed_Zz1)
yelp_academic_dataset_user_fixed_Zz1_pd = getDataFrameFromDict()
yelp_academic_dataset_user_fixed_Zz1_pd
```
